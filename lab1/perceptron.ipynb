# Simple perceptron

```python
class WightStabilizator:
    def __init__(self, y, e, t = 1):
        self.Alpha = 1
        self.Y = y
        self.E = e
        self.T = t
        
    def Rozenblad(self, weight, x0):
        return weight - self.Alpha * x0 * (self.Y - self.E)
    
    def Hebba(self, weight, x0):
        return weight * (1 - 0.2) - self.Alpha * x0 * (self.Y - self.E)
    
    def WidrowHoffa(self, weight, x0):
        alpha = self.Alpha / self.T
        return weight - alpha * (self.Y**self.T - self.E**self.T)* x0 **self.T
    
    def CorrectionsMistakes(self, weight, x0):
        return weight + 2*(self.Y - self.E)
    
```

```python
global T
T = 0
def sigmoid(x):
  return 0 if x <= T else 1
        
class Neuron:
    def __init__(self, weights, bias):
        self.Weights = weights
        self.Bias = bias
    
    def feedforward(self, inputs):
        self.LastInput = inputs
        total = np.dot(self.Weights, inputs) + self.Bias
        self.LastRezult = sigmoid(total)
        return sigmoid(total)
    
    
    def WeightCorrection(self, function):
        for i in range(len(self.Weights)):
                self.Weights[i] = function(self.Weights[i], self.LastInput[i])
                
        self.Bias = function(self.Bias, self.Bias)
                

    
```

# A

```python
class PerceptronA:
    def __init__(self):
        self.h1 = Neuron(np.random.rand(2), 0)
        self.h2 = Neuron(np.random.rand(2), 0)
        self.h3 = Neuron(np.random.rand(2), 0)
        
        self.r1 = Neuron(np.random.rand(3), 0)
    
    def fit(self, data, y):
        t = 1
        print('t =', t, 'Weigths =', self.h1.Weights,self.h2.Weights, self.h3.Weights, self.r1.Weights)
        flag = True
        while flag and t < 100:
            flag = False
            for i in range(len(data)):
                e = self.feedforward(data[i])
                if e != y[i]:
                    self.WeightCorrection(e, y[i])
                    t+=1
                    print('t = ', t, 'Weigths = ', self.h1.Weights,self.h2.Weights, self.h3.Weights, self.r1.Weights)
                    flag = True
    
    def predict(self, data):
        return self.feedforward(data)
        
    def feedforward(self, inputs):
        h1_res = self.h1.feedforward(inputs)
        h2_res = self.h2.feedforward(inputs)
        h3_res = self.h3.feedforward(inputs)
        
        return self.r1.feedforward(np.array([h1_res,h2_res,h3_res]))
    
    def WeightCorrection(self,y,e):
        stabil = WightStabilizator(y, e)
        self.r1.WeightCorrection(stabil.Rozenblad)
        self.h1.WeightCorrection(stabil.Rozenblad)
        self.h2.WeightCorrection(stabil.Rozenblad)
        self.h3.WeightCorrection(stabil.Rozenblad)
```

```python
perc = PerceptronA()
perc.fit(dataSet, Y)

print(Y)
for i in dataSet:
    print(perc.predict(i))
```

```python
perc = PerceptronB()
perc.fit(dataSet, Y)

print(Y)
for i in dataSet:
    print(perc.predict(i))
```
